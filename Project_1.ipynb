In the function we're creating, use variables to store rows from the original table, where:

  * The value in the `'city'` column is equal to the `city` parameter

Then, count the values in the `'user_id'` column in the resulting table. Save the result into a new variable. Return this variable from the function.

```python
# Verify the function procedure
track_list = df.loc[(df['city']=='Springfield') & (df['day']=='Monday')]
track_list_count = track_list['user_id'].count()
track_list_count
```

```python
# <creating the function number_tracks()>
# We will declare a function with two parameters: day=, city=.
# Make the track_list variable store rows from df, where
# the value in the 'day' column is equal to the day= parameter, and at the same time,
# the value in the 'city' column is equal to the city= parameter (apply sequential filtering
# with logical indexing).
# Make the track_list_count variable store the count of values in the 'user_id' column in track_list
# (find it using the count() method).
# Make sure the function you create yields a number: the value of track_list_count.

# The function calculates the tracks played for a specific city and day.
# First, it will fetch rows with the desired day from the table,
# then filter those rows with the desired city from the result,
# then find the count of 'user_id' values in the filtered table,
# then yield that count.
# To see what it yields, package the function call in print().

def number_tracks(datas, day, city):
    track_list = datas.loc[(datas['city']==city) & (datas['day']==day)]
    track_list_count = track_list['user_id'].count()
    track_list_count
    return track_list_count
```

Create a function `genre_weekday()` with four parameters:
* A table for data
* Day name
* Start time stamp, in the format 'hh:mm'
* End time stamp, in the format 'hh:mm'

The function should output information about the 15 most popular genres on a specific day during the period between the two time stamps.

```python
# Declare the genre_weekday() function with parameters day=, time1=, and time2=. The function should
# provide information about the most popular genres on a specific day and time:

# 1) Make the variable genre_df store the rows that meet the following conditions:
#   - the value in the 'day' column is equal to the argument day=
#   - the value in the 'time' column is greater than the argument time1=
#   - the value in the 'time' column is less than the argument time2=
#   Use sequential filtering with logical indexing.

# 2) Group genre_df by the 'genre' column, select one of its columns,
#    then use the count() method to find the number of entries for each
#    represented genre; store the resulting Series in the variable genre_df_count

# 3) Sort genre_df_count in descending order by frequency and store the result
#    in the variable genre_df_sorted

# 4) Return a Series object with the top 15 genre_df_sorted values - the 15 most
#    popular genres (on a specific day, within a specific time period)

# Function
def genre_weekday (dataframe, day, time1, time2):
    # Sequential filtering
    # genre_df will only store rows from df where 'day' is equal to the parameter day

    genre_df = dataframe.loc[dataframe['day']==day]

    # genre_df will only store rows from df where 'time' is less than time2
    genre_df = genre_df.loc[genre_df['time']<time2]

    # genre_df will only store rows from df where 'time' is greater than time1
    genre_df = genre_df.loc[genre_df['time']>time1]

    # group the filtered DataFrame by the column named genre, take the genre column, and find the count of rows for each genre using the count() method.
    genre_df_grouped = genre_df.groupby('genre')['user_id'].count()

    # We'll sort the results in descending order (so that the most popular genres are displayed earlier in the Series object).
    genre_df_sorted = genre_df_grouped.sort_values(ascending=False)

    # We'll generate a Series object that stores the 15 most popular genres on a specific day within a certain time frame.
    return genre_df_sorted[:15]
```

Compare the results from the `genre_weekday()` function for Springfield and Shelbyville on Monday morning (from 07:00 to 11:00) and on Friday evening (from 17:00 to 23:00):

```python
# calling the function for Monday morning in Springfield (use spr_general instead of the df table)
genre_weekday (dataframe=spr_general, day='Monday', time1='07.00', time2='11.00')
```

```python
# calling the function for Monday morning in Shelbyville (use shel_general instead of the df table)
genre_weekday (dataframe=shel_general, day='Monday', time1='07.00', time2='11.00')
```

```python
# calling the function for Friday night in Springfield
genre_weekday (dataframe=spr_general, day='Friday', time1='17:00', time2='23:00')
```

```python
# calling the function for Friday night in Shelbyville
genre_weekday (dataframe=shel_general, day='Friday', time1='17:00', time2='23:00')
```

**Conclusion**

After comparing the top 15 genres on Monday morning, we can draw the following conclusions:

1. Users from Springfield and Shelbyville listen to music of the same genre. The top five genres from both cities are the same, with only rock and electronic genres switching places.

2. In Springfield, the amount of missing values is significant, with the `'unknown'` value ranking tenth. This indicates that the missing values constitute a substantial proportion of the data, raising questions about the reliability of our conclusions.

For Friday night, the situation is similar. Individual genres vary, but overall, the top 15 genres for both cities are the same.

Thus, hypothesis two is partially confirmed:
* Users listen to the same music at the beginning and end of the week.
* There are no significant differences between Springfield and Shelbyville. In both cities, pop is the most popular genre.

However, the significance of the missing values raises doubts about these results. In Springfield, there are so many missing values that affect our top 15 genre results. Without these missing values, the results might differ.
